```python
import pandas as pd
College = pd.read_csv("Datasets/College.csv")
```

```python
College.head(5)
```

```python
College2 = pd.read_csv('Datasets/College.csv', index_col =0) #Set the first column as index, here the name
College2
```

```python
college3 = College.rename ({'Unnamed: 0': 'College '},axis =1) #axis=1 to look in the column names (0 for rows)
college3 = college3.set_index('College ') #equivalent method as college2
college=college3
```

```python
type(college) #It is a panda data frame
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc','Apps', 'Enroll']]); #The diagonal contains histograms
```

```python
from matplotlib.pyplot import subplots 
fig, ax = subplots(figsize=(6, 6)) 
college.boxplot('Outstate', by='Private', ax=ax); #Creating a boxplot of one variable against another means you are comparing the distribution of the first variable across different categories defined by the second variable.
```

```python
#pd.cut is a very useful function for data preprocessing and feature engineering in machine learning
#and data analysis. It allows you to convert continuous variables into categorical ones, which can be useful for certain types of analysis or modeling tasks. 
college['Elite'] = pd.cut(college['Top10perc'],[0,50,100],labels =['No', 'Yes'])#This create a new column in the college dataset
college['Elite']
```

```python
college['Elite'].value_counts()
```

```python
fig, ax = subplots(figsize=(6, 6)) 
college.boxplot('Outstate', by='Elite', ax=ax);
```

```python
fig, ax = subplots(2,2,figsize=(7, 7)) 
college.hist('Outstate', bins=10, ax=ax[0,0]);
college.hist('PhD', bins=10, ax=ax[0,1]);
college.hist('Expend', bins=10, ax=ax[1,0]);
college.hist('Personal', bins=10, ax=ax[1,1]);
```

```python
fig, ax = subplots(2,2,figsize=(7, 7)) 
college.hist('Outstate', bins=20, ax=ax[0,0]);
college.hist('PhD', bins=20, ax=ax[0,1]);
college.hist('Expend', bins=5, ax=ax[1,0]);
college.hist('Personal', bins=5, ax=ax[1,1]);
```

#### Some additional exploration

```python
college['Accept'].describe()
```

```python
college['AcceptanceRate']=college['Accept']/college['Apps']
```

```python
college['EnrollmentRate']=college['Enroll']/college['Accept']
```

```python
college['AcceptanceRate'].describe()
```

```python
college['AcceptanceRateCategorical']=pd.cut(college['AcceptanceRate'],bins=[0,0.67,0.77,0.84,1], labels=False) #Dividing Accept category in subgroups
```

```python
fig, ax = subplots(figsize=(6, 6)) 
college.boxplot("EnrollmentRate", by="AcceptanceRateCategorical",ax=ax);
```

```python
import pandas as pd
Auto = pd.read_csv('Datasets/Auto.csv', na_values=['?'],index_col=8) #index_col to set a column as index
Auto=Auto.dropna()  #Drop the rows with a missing value
Auto
```

```python
Auto.describe() #All the values are quantitative except origin (cylinders could be considered as qualitative since there are not a lot of values)
```

```python
Auto.max()-Auto.min()
```

```python
AutoModif=pd.concat([Auto.iloc[:10],Auto.iloc[85:]])  #delete 10th to 85th observation
# Normally you could do something like Auto.drop(Auto.index[10:86]) but since the index are not unique, it deletes the cars that have the same name
```

```python
AutoModif.describe()
```

```python
Auto.describe()
```

```python
pd.plotting.scatter_matrix(Auto);
```

```python
from matplotlib.pyplot import subplots 
fig, ax = subplots(2,2,figsize=(9, 9)) 
Auto.boxplot('horsepower', by='cylinders', ax=ax[0,0]); 
Auto.boxplot('acceleration', by='cylinders', ax=ax[0,1]);
Auto.boxplot('weight', by='cylinders', ax=ax[1,0]);
Auto.boxplot('mpg', by='cylinders', ax=ax[1,1]);
```

```python
from matplotlib.pyplot import subplots #We can predict mpg by horsepower, weight and displacement. We can really see a correlation between those variabkes and mpg
fig, ax = subplots(4,2,figsize=(12, 12)) 
Auto.plot('horsepower','mpg',marker='o', kind='scatter', ax=ax[0,0]); 
Auto.plot('acceleration', 'mpg',marker='o', kind='scatter', ax=ax[0,1]);
Auto.plot('weight','mpg',marker='o', kind='scatter', ax=ax[1,0]);
Auto.plot('year', 'mpg',marker='o', kind='scatter',ax=ax[1,1]);
Auto.plot('cylinders', 'mpg',marker='o', kind='scatter',ax=ax[2,0]);
Auto.plot('displacement', 'mpg',marker='o', kind='scatter',ax=ax[2,1]);
Auto.plot('origin', 'mpg',marker='o', kind='scatter',ax=ax[3,0]);
Auto.hist('mpg',ax=ax[3,1]);
```
